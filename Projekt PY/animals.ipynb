```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
animal_df = pd.read_csv('static/animals3.csv', delimiter=';')
animal_df.head(n=51)
```

```python
type_df = animal_df['Animal Type'].value_counts().reset_index()

type_df.head()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(type_df['count'], labels=type_df['Animal Type'], autopct='%1.1f%%', startangle=90)

# Add title
plt.title('Animal Type Distribution')

# Create an interactive pie chart
fig = px.pie(type_df, names='Animal Type', values='count', title='Animal Type Distribution')
fig = px.pie(
    type_df, 
    names='Animal Type', 
    values='count', 
    title='Animal Type Distribution',
    color='Animal Type',  
    color_discrete_map={"Dog": "#a5ec91", "Cat": "#29920c", "Rabbit": "#76f952", "Guinea Pig": "#59a742"},)
    

# Save the figure as an HTML file
fig.write_html('animal_type_distribution.html')
plt.close()

```
